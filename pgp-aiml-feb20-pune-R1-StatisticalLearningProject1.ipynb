### 1) Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

```python
# Read data from Laliga.csv
team_data = pd.read_csv("Laliga.csv", skip_blank_lines=True, header=1)
team_data.head()
```

```python
# Replace - with 0
team_data.replace('-', 0, inplace=True)
team_data.head(10)
```

### 2) Print all the teams which have started playing between 1930-1980.

```python
# defined fuction that will return Debut year
def get_team_start_year(year):
    start_year = int(year[:4])
    #print(start_year, type(start_year))
    return start_year

# Check Debut year for each team and print it
for i in team_data.index:
    debut = get_team_start_year(team_data['Debut'][i])
    if debut >=1930 | debut <=1980:
        print('-----------------------------------------------------------------------')
        print("Team :", team_data['Team'][i], '\nDebut Year :', team_data['Debut'][i])
```

### 3) Print the list of teams which came Top 5 in terms of points

```python
# Convert data type of point column to numeric
team_data['Points'] = team_data['Points'].apply(pd.to_numeric)
```

```python
# Sort data frame and print top 5 teams
team_data.sort_values(by=['Points'], ascending = False).Team[:5]
```

### 4) Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

```python
# Define function for Goal difference count and store it in to same data frame with new column
def Goal_diff_count(df):
    Goal_diff = []
    for i in df.index:
        Goal_diff.append(df.GoalsFor[i] - df.GoalsAgainst[i])
        
    df['GoalDiff'] = Goal_diff
    return df    

# convert data type for required columns
team_data['GoalsFor'] = team_data['GoalsFor'].apply(pd.to_numeric)
team_data['GoalsAgainst'] = team_data['GoalsAgainst'].apply(pd.to_numeric)

# Call the function to create new column with goal difference
Goal_diff_count(team_data)

# Print team with Max and min goal difference
print("Max :", team_data['Team'][team_data['GoalDiff'].idxmax()], ",", team_data['GoalDiff'].max())
print("Min :", team_data['Team'][team_data['GoalDiff'].idxmin()], ",", team_data['GoalDiff'].min())
```

### 5) Create a new column with the name “Winning Percent” and append it to the data set

```python
# Define fuction for winning percentage, and store it in same dataframe with new column
def winning_percent(df):
    win_per = []
    for i in df.index:
        if df.GamesPlayed[i] <= 0:
            win_per.append(0)
        else:
            win_per.append((df.GamesWon[i] / df.GamesPlayed[i]) * 100)
        
    df['WinninPercent'] = win_per
    return df    

team_data['GamesWon'] = team_data['GamesWon'].apply(pd.to_numeric)
team_data['GamesPlayed'] = team_data['GamesPlayed'].apply(pd.to_numeric)

# Call the funcation
winning_percent(team_data)

# print top 5 team with most winning percentage
team_data.sort_values(by=['WinninPercent'], ascending = False)[:5]
```

### 6) Group teams based on their “Best position” and print the sum of their points for all positions
